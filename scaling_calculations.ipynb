Scaling

```python
# General constants
g = 9.81 # m/s^2
rho = 1025 # kg/m^3
rho_steen = 2650 # kg/m^3
nu = 1.33e-6 # m^2/s
bodemdiepte = 40 # mMSL
Kv = 1.1 # Velocity coefficient
Delta_v = (rho_steen-rho) / rho # Dimensionless
```

```python
h_p = bodemdiepte  # Prototype water depth (m)
h_m = 1.2         # Model water depth (m)

u_p = 2.0  # Prototype velocity (m/s) AANPASSEN
d_p = 0.3 # Prototype rock size (m) AANPASSEN

H_p = 0.5 # Prototype wave height (m) AANPASSEN

Tp_p = 5.0 # Prototype wave period (s) AANPASSEN
```

```python
n_L = h_p / h_m
print(f'n_L: {n_L:.3f} [-] scaling factor for water depth') #based on flume 1.2 m deep

n_delta = 1 #assumed that the density difference is the same in model and prototype
n_d = n_L / n_delta #scaling rule for the rock diameter
print(f'n_d: {n_d:.3f} [-] scaling factor for rock diameter')

print()

u_m = u_p / n_L**0.5
print(f'u_m: {u_m:.3f} [m/s] current velocity model')

d_m = d_p / n_d
print(f'd_m: {d_m:.3f} [m] rock diameter model')

H_m = H_p / n_L
print(f'H_m: {H_m:.3f} [m] wave height model')

Tp_m = Tp_p / n_L**0.5
print(f'Tp_m: {Tp_m:.3f} [s] wave period model')

delta = (rho_steen / rho) - 1
print(f'delta: {delta:.3f} [-] relative density difference')

print()

Re_p = (u_p * d_p) / nu
print(f'Re_p: {Re_p:.3f} [-] Reynolds number prototype')

Re_m = (u_m * d_m) / nu
print(f'Re_m: {Re_m:.3f} [-] Reynolds number model')
```
