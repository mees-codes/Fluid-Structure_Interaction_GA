```python
def int_stability(d60, d10):
    Int_stability_treshold = 10
    stability = d60 / d10
    if stability < Int_stability_treshold:
        return print("Internal stability satisfied with a value of", stability)
    else:
        return print("Internal stability NOT satisfied with a value of", stability)

def Stability_check(d15f, d85b):
    Stability_treshold = 5
    stability = d15f / d85b
    if stability < Stability_treshold:
        return print("Stability condition satisfied with a value of", stability)
    else:
        return print("Stability condition NOT satisfied with a value of", stability)

def permeability_check(d15f, d15b):
    Permeability_treshold = 5
    permeability = d15f / d15b
    if permeability > Permeability_treshold:
        return print("Permeability condition satisfied with a value of", permeability)
    else:
        return print("Permeability condition NOT satisfied with a value of", permeability)
    
```
