```python
import numpy as np

u_c = 2
delta = 2650/1025 - 1
phi_c = 0.04
C = 50
K_v = 1.55

for i in range(5):
    d_n50 = u_c**2 / (delta * phi_c * C**2)
    print(f'Iteration {i+1}: d_n50 = {d_n50:.3f} m')
    C = 18.0 * np.log10(12.0 * 40 / (2 * d_n50))

d_n50 = K_v**2 * u_c**2 / (delta * phi_c * C**2)

print(f'Required d_n50: {d_n50*100:.3f} cm')
```
